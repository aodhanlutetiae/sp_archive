## Query for first 1000 files

```python
# empty list to hold the dictionaries we'll get for each file

# --- NOTE. AS LONG AS THE KERNEL IS LIVE THIS CAN BE COMMENTED OUT AND THE CODE RERUN BY ADJUSTING THE INTEGER THAT
# --- THE LOOP STARTS ON (CURRENTLY 1). IF IT'S LEFT IN THE CODE THE LIST WILL EMPTY AND START AT THE BEGINNING.

# big_list = []

# this code seems to work for all the queries
codice = '85df5s%244Kue%C3%B1wQw5O2p4J1G9'

# this number refers to the file number that changes with each file / url
nb = 1

for x in range(1000):
    file_nb = str(nb)
    query = f'https://dedalo4.bancmemorial.extranet.gencat.cat/dedalo/lib/dedalo/publication/server_api/v1/json/records?code={codice}&lang=lg-cat&table=deportats&count=true&section_id={file_nb}&resolve_portal=true&resolve_portals_custom=%7B%22deportat%22:%22informant%22,%22deportat.biography%22:%22biografia%22,%22exercit_frances%22:%22deportats_treballs_exercit_frances%22,%22camps_francesos%22:%22deportats_camps_francesos%22,%22camps_concentracio%22:%22deportats_camps_concentracio%22,%22consultes_arxius%22:%22deportats_consultes_arxius%22,%22kommando_extern%22:%22deportats_kommando_extern%22,%22publicacions%22:%22referencia_bibliografica%22,%22publicacio_web%22:%22referencia_bibliografica%22,%22empresonament%22:%22deportats_empresonament%22,%22tren_salida%22:%22deportats_trens%22,%22organizacio_todt%22:%22deportats_todt%22,%22indemnitzacions%22:%22deportats_indemnitzacions%22%7D&ar_fields=section_id,data_mod,deportat,exercit_frances,camps_francesos,camps_concentracio,consultes_arxius,kommando_extern,publicacions,publicacio_web,situacio_deportat,data_situacio,ref_lloc_situacio,unitat_militar,frontera,destinacio,graduacio_militar,data_pas_per_frontera,resistencia,lloc_lluita,cut_lloc_lluita,lloc_lluita_original,lloc_detencio,ref_lloc_detencio,lloc_detencio_original,data_detencio,empresonament,tren_salida,organizacio_todt,ref_observacions,ref_repatriacio_franca,data_repatriacio_franca,data_tornarda_espanya,ref_lloc_residencia_posterior,ref_emigracio,indemnitzacions,obs_deportacio'
    req = requests.get(query, timeout = 5)
    response = req.json()
    
    if req.status_code != 200:
        data = 'CALL FAILED HERE'
    if len(response['result']) == 0:
        data = 'NO DATA IN THIS CALL'
    else:
        data = response['result'][0]
    big_list.append(data)

    if nb%25 == 0:
        print(nb)
    nb = nb + 1
    time.sleep(random.randint(1, 3))
```

```python
# print the relevant fields for an individual

individual = big_list_imp[12]

sec_id = individual['section_id']
url = f'https://banc.memoria.gencat.cat/ca/app/#/result/deportats/{sec_id}'
print(url)
print('-------')

perso = individual['deportat'][0]
print(perso['name'])
print(perso['surname'])
print(perso['birthdate'])
print(perso['birthplace'])
print(perso['dead_date'])
print(perso['cut_deadplace'])
print('-------')

print(individual['situacio_deportat'])
print(individual['data_situacio'])
print(individual['ref_lloc_situacio'])
print('-----')

prison = individual['empresonament'][0]
print(prison['tipus_preso'])
print(prison['nom_preso'])
print(prison['data_sortida'])
print(prison['cut_localitat'])
print(prison['num_matricula'])

tren = individual['tren_salida'][0]
print(tren['nom_estacio'])
print(tren['ref_localitat'])
print('-----')

camp = individual['camps_concentracio'][0]
print(camp['ref_camp'])
print(camp['data_entrada'])
print(camp['num_matricula'])

print('-----')
end = individual['kommando_extern'][0]
print(end['ref_kommando_extern'])
print(end['data_entrada'])
print(end['num_matricula'])
```
